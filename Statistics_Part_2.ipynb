Q1.What is hypothesis testing in statistics?

  ->> Hypothesis Testing in Statistics Hypothesis testing is a method used in statistics to make decisions or inferences about a population based on sample data. It’s like putting an assumption on trial to see if the evidence backs it up.

Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

  ->> The Null Hypothesis (H₀) The null hypothesis is a fundamental concept in statistics used when conducting a hypothesis test. It’s the starting point — the assumption that nothing has changed, nothing is different, or there’s no effect.

->>
Null Hypothesis (H₀)
- Represents a baseline or default assumption

- Suggests no effect, no difference, or no relationship

- It's the idea you're trying to test against

- Example: “This new drug has no effect on blood pressure.”

🟠 Alternative Hypothesis (H₁ or Ha)
- Represents a new claim or research assumption

- Suggests an effect, a difference, or a relationship

- It's what you're trying to find evidence for

- Example: “This new drug does affect blood pressure.”


Q3. What is the significance level in hypothesis testing, and why is it important?

  ->> Significance Level (α) in Hypothesis Testing The significance level, denoted by α (alpha), is a threshold used to determine how strong the evidence must be before rejecting the null hypothesis.

Why It’s Important

- Controls Errors: It protects your conclusion from being unreliable. A lower α means you're being more cautious.

- Shapes the Outcome: The smaller the α, the stronger the evidence must be to reject H₀.

- Used for Comparison: The p-value from your test is compared against α to decide whether to reject the null hypothesis.

Q4.What does a P-value represent in hypothesis testing?

  ->> P-value in Hypothesis Testing The p-value is a number that tells you how likely your observed results are, assuming the null hypothesis is true. It helps you decide whether to reject the null hypothesis.

Q5.How do you interpret the P-value in hypothesis testing?

  ->> Interpreting the P-value: A Quick Guide to Meaningful Conclusions

The p-value plays a central role in hypothesis testing—it tells you how surprising your data is if the null hypothesis (H₀) were true.

Q6.What are Type 1 and Type 2 errors in hypothesis testing?

  ->>
❌ Type I Error (False Positive)

- Occurs when you reject the null hypothesis (H₀) when it's actually true

- You think there’s an effect or difference—but it’s just random chance

- Probability of making this error = significance level (α), often set at 0.05

❌ Type II Error (False Negative)

- Happens when you fail to reject the null hypothesis when the alternative is actually true

- You miss a real effect or difference because the evidence wasn’t strong enough

- Probability of making this error = β (beta). The complement, 1 - β, is called power.

Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

  ->> 🍃 One-Tailed Test

- Tests for an effect in one specific direction

- You’re asking: "Is it greater than or less than?"

Example:

  - H₀: Lime formate concentration ≤ 5 mg/L

  - H₁: Lime formate concentration > 5 mg/L

Use when you're only interested in improvement or increase/decrease—not both

🍃 Two-Tailed Test

- Tests for any difference — higher or lower

- You’re asking: "Is it different—in either direction?"

Example:

  - H₀: Lime formate concentration = 5 mg/L

  -H₁: Lime formate concentration ≠ 5 mg/L

Use when you’re checking for any deviation, not just more or less

Q8.What is the Z-test, and when is it used in hypothesis testing?

  ->> The Z-test is a statistical test used to determine whether there's a significant difference between a sample statistic and a known population parameter. It’s ideal when certain conditions are met and helps you decide if your results are due to chance or reflect a real effect.

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

  ->> 🔢 Formula to Calculate Z-score
For a sample mean
              z = (X - μ) / σ
where,

z = Z-Score
X = Value of Element
μ = Population Mean
σ = Population Standard Deviation

🧪 What the Z-score Represents

- A Z-score of 0 → sample mean equals population mean

- A positive Z-score → sample mean is above population mean

- A negative Z-score → sample mean is below population mean

- The higher the absolute value, the more extreme the difference

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

  ->> The T-distribution, also known as Student’s t-distribution, is a probability distribution that’s similar to the normal distribution — but with fatter tails. This helps account for the extra uncertainty that comes with smaller sample sizes or unknown population parameters.

Q11.What is the difference between a Z-test and a T-test?

  ->> 🧮 Z-Test
- Used when the population standard deviation (σ) is known

- Best for large sample sizes (typically n ≥ 30)

- Assumes data follows a normal distribution

- Based on the standard normal (Z) distribution

🔹 Example: You’re testing whether the average lime formate level differs from a known value, using a sample of 100 and knowing the population σ → Use a Z-test

📐 T-Test
- Used when population σ is unknown and must be estimated from the sample

- Ideal for small samples (typically n < 30)

- Accounts for more variability using the T-distribution

- Has heavier tails, which gives more room for error

🔹 Example: You’re comparing two teaching methods with just 15 students in each group → Use a T-test

Q12.What is the T-test, and how is it used in hypothesis testing?

  ->> T-Test in Hypothesis Testing: Finding Meaning in Small Samples

A T-test is a statistical method used to determine if there’s a significant difference between means, often applied when:

- You have a small sample size

- The population standard deviation is unknown

🧠 How It’s Used in Hypothesis Testing
1.State your hypotheses

  - Null (H₀): No difference between means

  - Alternative (H₁): There is a difference

2.Choose significance level (α) — often 0.05

3.Calculate t-statistic and compare to critical value (or find p-value)

4.Make a decision

  - If |t| > critical value → Reject H₀

  - If p-value ≤ α → Reject H₀

Q13. What is the relationship between Z-test and T-test in hypothesis testing?

  ->>
- Both compare a sample mean to a population mean (or another sample mean)

- Both rely on a test statistic that reflects how far your sample result is from expectation

- Both help decide whether to reject or fail to reject the null hypothesis (H₀)

Q14. What is a confidence interval, and how is it used to interpret statistical results?

  ->> Confidence Interval (CI): Measuring Uncertainty with Precision

A confidence interval gives you a range of values that’s likely to contain a population parameter (like a mean or proportion), based on your sample data. It adds context to your estimate by showing how reliable or precise it is.

🧠 How It’s Used in Interpretation
- A narrow CI means your estimate is precise.

- A wide CI means there’s more uncertainty in your estimate.

- If the CI for a difference includes 0, it suggests the difference may not be statistically significant.

- CIs complement p-values by providing estimation, not just a yes/no decision.

Q15.What is the margin of error, and how does it affect the confidence interval?

  ->> Margin of Error: Putting a Range Around Your Estimates

The margin of error (MOE) is a number that shows how much your sample estimate might differ from the true population value due to chance alone. It's the "wiggle room" around your estimate, and it's a key part of building a confidence interval (CI).

🧠 How MOE Affects the Confidence Interval
- Larger MOE → wider confidence interval → less precise

-Smaller MOE → narrower interval → more precise

Q16.How is Bayes' Theorem used in statistics, and what is its significance?

  ->> Bayes’ Theorem in Statistics: Turning Prior Beliefs into Updated Reality

Bayes’ Theorem is a cornerstone of statistical inference that helps us update our understanding of a situation as new evidence rolls in. It's all about shifting probabilities based on fresh insights.

Q17.What is the Chi-square distribution, and when is it used?

  ->> Chi-square Distribution: A Tool for Testing Categorical Data

The Chi-square (χ²) distribution is a right-skewed probability distribution that arises frequently in statistical hypothesis testing, especially when working with counts and frequencies rather than averages.

Q18. What is the Chi-square goodness of fit test, and how is it applied?

  ->> Chi-square Goodness-of-Fit Test: Checking How Well Observed Matches Expected

The Chi-square goodness-of-fit test helps you determine whether a set of observed categorical data fits a specific expected distribution

📋 Steps to Apply the Test
1.State Hypotheses

  - H₁: The observed frequencies do not follow the expected distribution.

2.Collect Data

- Observe frequencies for each category.

 -If χ² ≤ critical value → Fail to reject H₀

Q19. What is the F-distribution, and when is it used in hypothesis testing?

  ->> F-distribution: Comparing Variability with Confidence

The F-distribution is a key player in statistics used when you're comparing variances across groups. It's a right-skewed distribution that becomes more symmetric as sample sizes grow. Its shape depends on two sets of degrees of freedom: one for the numerator and one for the denominator.

Q20.What is an ANOVA test, and what are its assumptions?

  ->>
It examines variability between group means compared to variability within the groups:

- H₀ (Null Hypothesis): All group means are equal

- H₁ (Alternative Hypothesis): At least one group mean is different

📌 Key Assumptions of ANOVA
These must be met for valid results:

1.Independence of observations

- Each data point is unrelated to the others

2.Normality

- The data in each group should be approximately normally distributed

3.Homogeneity of variances

- All groups should have roughly equal variances (also called homoscedasticity)

Q21. What are the different types of ANOVA tests?

  ->>
1.One-Way ANOVA
- Purpose: Tests for differences in means across three or more independent groups using one factor

2.Two-Way ANOVA
- Purpose: Tests the effect of two independent variables (factors) and their interaction on one dependent variable

3.Repeated Measures ANOVA
- Purpose: Tests differences when the same subjects are measured under multiple conditions or time points

4.Mixed-Design ANOVA (Split-Plot)
- Purpose: Combines between-subjects factors (like different groups) and within-subjects factors (like repeated measures)

Q22. What is the F-test, and how does it relate to hypothesis testing?

  ->>
- The test statistic follows the F-distribution, which is right-skewed, and depends on two degrees of freedom: one for the numerator and one for the denominator.

- The larger the F-value, the greater the difference in variability between groups.

How It Relates to Hypothesis Testing

- Null hypothesis (H₀): No difference between variances (or models have no effect)

- Alternative hypothesis (H₁): A difference exists
  You calculate the F-statistic, then:

- Compare it to a critical value from the F-distribution table

- Or use a p-value to assess significance

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the distribution
low = 1   # minimum value
high = 7  # maximum value (exclusive, so this gives values from 1 to 6)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Values in the support
x = np.arange(low, high)

# Compute PMF
pmf = dist.pmf(x)

# Plot the PMF
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution (Dice Roll)")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("Bernoulli outcome must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("PDF at x=1, p=0.7:", bernoulli_pdf(1, 0.7))  # Should return 0.7
print("PDF at x=0, p=0.7:", bernoulli_pdf(0, 0.7))  # Should return 0.3
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters
low = 1       # Minimum value (inclusive)
high = 7      # Maximum value (exclusive): values 1 to 6

# Create a discrete uniform distribution
dist = randint(low, high)

# Support of the distribution
x = np.arange(low, high)

# Compute the CDF
cdf = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf, where='post', color='purple', linewidth=2)
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xticks(x)
plt.grid(alpha=0.4)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    data (list or numpy array): Dataset to process
    """
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

# Example usage:
data = [60, 72, 75, 85, 90, 65, 70, 80, 95, 100]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

# Given values
population_mean = 1000       # Claimed mean (μ)
sample_mean = 980            # Observed sample mean (x̄)
sample_size = 50             # Number of observations (n)
population_std = 100         # Known population standard deviation (σ)
alpha = 0.05                 # Significance level

# Calculate Z-statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate critical value for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print(f"Z-score: {z_score:.4f}")
print(f"Critical Z-value: ±{z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: The sample provides enough evidence to refute the claim.")
else:
    print("Fail to reject the null hypothesis: Not enough evidence against the claim.")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats

# Example dataset
data = [72, 75, 78, 71, 69, 77, 73, 70, 74, 76]

# Step 1: Compute sample statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(data)

# Step 2: Compute standard error and t critical value
confidence_level = 0.95
alpha = 1 - confidence_level
t_crit = stats.t.ppf(1 - alpha/2, df=n - 1)
margin_of_error = t_crit * (std_dev / np.sqrt(n))

# Step 3: Calculate confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Step 4: Display the results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(mean=0, std_dev=1, size=1000):
    # Generate dataset from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data, bias=False)
    kurt = kurtosis(data, bias=False)  # Excess kurtosis (normal = 0)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (excess): {kurt:.4f}")
    return skewness, kurt

# Example usage
analyze_distribution(mean=100, std_dev=15, size=5000)
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference.")

# Example usage
sample = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99,
          105, 109, 109, 109, 110, 112, 112, 113, 114, 115]
population_mean = 100
population_std = 15

one_sample_z_test(sample, population_mean, population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=105, scale=10, size=50)  # mean=105, std=10
population_mean = 100
population_std = 10

# Z-test calculation
sample_mean = np.mean(sample)
sample_size = len(sample)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label='Rejection Region')

# Plot Z-statistic
plt.axvline(z_stat, color='green', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')
plt.axvline(-z_critical, color='black', linestyle=':', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':')

plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05, sample_size=30):
    # Critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Standard error
    se = sigma / np.sqrt(sample_size)

    # Define range for sample means
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    # Null and alternative distributions
    null_dist = norm.pdf(x, loc=mu_null, scale=se)
    alt_dist = norm.pdf(x, loc=mu_alt, scale=se)

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H₁)', color='green')

    # Shade Type I error regions
    plt.fill_between(x, 0, null_dist, where=(x <= mu_null - z_critical*se) | (x >= mu_null + z_critical*se),
                     color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region
    plt.fill_between(x, 0, alt_dist, where=(x > mu_null - z_critical*se) & (x < mu_null + z_critical*se),
                     color='orange', alpha=0.4, label='Type II Error (β)')

    # Labels and legend
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05, sample_size=30)
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    # Convert to NumPy array
    data = np.array(data)

    # Sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    df = n - 1  # Degrees of freedom

    # t critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    # Margin of error
    margin = t_crit * std_err

    # Confidence interval
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return (lower, upper)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    std_err = std_dev / np.sqrt(n)  # Standard error
    df = n - 1  # Degrees of freedom

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    # Margin of error
    moe = t_crit * std_err

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {int(confidence * 100)}%")
    print(f"Margin of Error: ±{moe:.2f}")
    return moe

# Example usage
sample_data = [25, 28, 30, 32, 29, 27, 31, 26, 28, 30]
margin_of_error(sample_data, confidence=0.95)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np
import matplotlib.pyplot as plt

# Prior: uniform belief over bias (theta)
theta = np.linspace(0, 1, 1000)
prior = np.ones_like(theta)  # Uniform prior

# Observed data: 7 heads out of 10 tosses
heads = 7
tails = 3

# Likelihood: Binomial distribution
likelihood = theta**heads * (1 - theta)**tails

# Posterior: Bayes' rule (unnormalized)
posterior_unnormalized = likelihood * prior
posterior = posterior_unnormalized / np.sum(posterior_unnormalized)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(theta, posterior, color='purple')
plt.title('Posterior Distribution of Coin Bias (θ)')
plt.xlabel('θ (Probability of Heads)')
plt.ylabel('Posterior Probability')
plt.grid(True)
plt.show()
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table: Gender vs Preference
# Rows: Male, Female
# Columns: Likes A, Likes B, Likes C
data = np.array([[120, 90, 40],
                 [110, 95, 45]])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Output results
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: Variables are associated.")
else:
    print("Fail to reject the null hypothesis: Variables are independent.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data


import numpy as np
import pandas as pd

# Example observed data: 2 rows (e.g., Gender), 3 columns (e.g., Product Preference)
observed = np.array([[120, 90, 40],
                     [110, 95, 45]])

# Convert to DataFrame for easier manipulation
df_obs = pd.DataFrame(observed, index=['Male', 'Female'], columns=['A', 'B', 'C'])

# Calculate row totals, column totals, and grand total
row_totals = df_obs.sum(axis=1)
col_totals = df_obs.sum(axis=0)
grand_total = df_obs.values.sum()

# Calculate expected frequencies
expected = np.outer(row_totals, col_totals) / grand_total
df_exp = pd.DataFrame(expected, index=df_obs.index, columns=df_obs.columns)

# Display results
print("Observed Frequencies:\n", df_obs)
print("\nExpected Frequencies:\n", df_exp.round(2))
```
